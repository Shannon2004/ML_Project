```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import warnings

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/kaggle/input/lend-or-lose/Processed_train.csv')
```

```python
data
```

```python
X = data.drop(['Default'],axis=1)
X
```

```python
Y = data['Default']
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

## Standardizing the data

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=[X_train.columns])
X_train_scaled_df
```

```python
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=[X_train.columns])
X_test_scaled_df
```

## Building Neural Network with two hidden layers. Using the relu activation function in the hidden layers. Using dropout for preventing overfitting. Using batch Normalization to ensure standardization. Adam Optimizer used to update weights.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization



model1 = Sequential()
model1.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model1.add(BatchNormalization())
model1.add(Dropout(0.5))
model1.add(Dense(32, activation='relu'))
model1.add(BatchNormalization())
model1.add(Dropout(0.3))
model1.add(Dense(1, activation='sigmoid'))

model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model1.summary()
```

## Stop when val loss stops decreasing for 25 epochs. Monitor the val loss and save the weights which gave the best val loss.

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Early stopping to avoid overfitting
early_stopping = EarlyStopping(monitor='val_loss',patience=25, restore_best_weights=True)

# Model checkpoint to save the best model
model_checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)
```

```python
# Train the model
history = model1.fit(X_train_scaled_df, y_train, epochs=100, batch_size=64, validation_split=0.2, 
                    callbacks=[early_stopping, model_checkpoint])
```

```python
model1.load_weights('best_model.keras')
```

```python
loss, accuracy = model1.evaluate(X_test_scaled_df, y_test)
print(f'Test Accuracy: {accuracy * 100:.2f}%')
```

```python
predictions1 = model1.predict(X_test_scaled_df)
predictions1 = (predictions1 > 0.5).astype(int)
```

```python
pred_1 = pd.DataFrame(predictions1)
pred_1.value_counts()
```

## Predict on test data

```python
sub_data = pd.read_csv('/kaggle/input/lend-or-lose/Processed_test.csv')
test_sub = sub_data.drop(['LoanID'],axis=1)
test_sub
```

```python
test_sub_scaled = scaler.transform(test_sub)
test_sub_scaled_df = pd.DataFrame(test_sub_scaled, columns=[test_sub.columns])
test_sub_scaled_df
```
